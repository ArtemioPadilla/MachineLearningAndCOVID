```python
raw_data = pd.read_excel(path,sheet_name=3)
raw_data = raw_data.merge(continents, on="Country")
```

```python
numeric = raw_data.loc[:,raw_data.columns[raw_data.dtypes==float]]
for col in numeric.columns:
    numeric[col].fillna(numeric[col].mean(),inplace=True)
numeric.dropna(inplace=True,axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

std = StandardScaler()
std_numeric = std.fit_transform(numeric)

```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering
from sklearn.metrics import silhouette_samples, silhouette_score

best_score = 0
print("Agglomerative Clustering")
print(" Cosine distance")
for n_components in range(2,50):
    pca = PCA(n_components=n_components)
    pca_numeric = pca.fit_transform(std_numeric)

    for i in range(2,20):
        for linkage in ["single","average","complete"]:    
            cluster = AgglomerativeClustering(i,affinity="cosine", linkage=linkage)
            clusters = cluster.fit_predict(pca_numeric)
            silhouette_avg = silhouette_score(pca_numeric, clusters, metric="cosine")
            if silhouette_avg> best_score:
                best_score = silhouette_avg
                print(n_components, i, silhouette_avg, linkage)


best_score = 0
print(" Euclidean distance")
for n_components in range(2,50):
    pca = PCA(n_components=n_components)
    pca_numeric = pca.fit_transform(std_numeric)

    for i in range(2,20):
        for linkage in ["single","average","complete","ward"]:    
            cluster = AgglomerativeClustering(i,affinity="euclidean", linkage=linkage)
            clusters = cluster.fit_predict(pca_numeric)
            silhouette_avg = silhouette_score(pca_numeric, clusters, metric="euclidean")
            if silhouette_avg> best_score:
                best_score = silhouette_avg
                print(n_components, i, silhouette_avg, linkage)

best_score = 0
print("KMeans Clustering")
print(" Euclidean distance")
for n_components in range(2,50):
    pca = PCA(n_components=n_components)
    pca_numeric = pca.fit_transform(std_numeric)

    for i in range(2,20): 
            cluster = KMeans(i)
            clusters = cluster.fit_predict(pca_numeric)
            silhouette_avg = silhouette_score(pca_numeric, clusters, metric="euclidean")
            if silhouette_avg> best_score:
                best_score = silhouette_avg
                print(n_components, i, silhouette_avg, linkage)
```

The best cluster is given by:

- PCA n_components = 2
- Single linkage on Agglomerative clustering with two clusters
- Comparing by thge cosine distance

```python
from sklearn import tree
def tree_classifier(X_train, y_train, X_test, y_test, max_depth=100, min_samples_leaf=2):

    dt = tree.DecisionTreeClassifier(max_depth=max_depth, min_samples_leaf=min_samples_leaf , random_state=43)
    acc = dt.fit(X_train, y_train).score(X_test, y_test)
    
    return dt, acc

def get_f1_score(clf, X_test, y_test):
  y_pred = clf.predict(X_test)
  return f1_score(y_test, y_pred)

def get_precision_score(clf, X_test, y_test):
  y_pred = clf.predict(X_test)
  return precision_score(y_test, y_pred)

def get_sensibility(clf, X_test, y_test):
  y_pred = clf.predict(X_test)
  return recall_score(y_test, y_pred)
```
